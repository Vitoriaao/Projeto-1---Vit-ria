```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
```

## Leitura dos arquivos em Excel:

#### IDH

Indice de desenvolvimento humano:

```python
idh = pd.read_excel('Indicator_HDI.xlsx')
```

#### HIV

Indice de adultos(15-49) com HIV em %:

```python
hiv = pd.read_excel('Indicator_hiv.xlsx')
```

#### AID

Indice de assistência oficial ao desenvolvimento recebida da população pelo estado %: 

```python
aid = pd.read_excel('indicator_aid received % of gni.xlsx')
```

## Selecionando os anos de 2000 e 2006:

```python
inco2000 = idh.loc[:, ['HDI', 2000]]
hiv2000 = hiv.loc[:, ['Estimated HIV Prevalence% - (Ages 15-49)', 2000]]
aid2000 = aid.loc[:, ['Aid received % of GNI', '2000']]
```

```python
inco2006 = idh.loc[:, ['HDI', 2006]]
hiv2006 = hiv.loc[:, ['Estimated HIV Prevalence% - (Ages 15-49)', 2006]]
aid2006 = aid.loc[:, ['Aid received % of GNI', '2006']]
```

## Mudando os indices:

Ano de 2000:

```python
inco_idx2 = inco2000.set_index('HDI')
hiv_idx2= hiv2000.set_index('Estimated HIV Prevalence% - (Ages 15-49)')
aid_idx2 = aid2000.set_index('Aid received % of GNI')
```

Ano de 2006:

```python
inco_idx = inco2006.set_index('HDI')
hiv_idx = hiv2006.set_index('Estimated HIV Prevalence% - (Ages 15-49)')
aid_idx = aid2006.set_index('Aid received % of GNI')
```

## Renomeando as duas primeiras colunas:

Ao renomear as colunas, foi colocado os três indices em data frames diferentes para cada ano.

Ano de 2000:

```python
df2 = inco_idx2.join(hiv_idx2, how='inner', lsuffix='_idh', rsuffix='_hiv')
```

Ano de 2006:

```python
df = inco_idx.join(hiv_idx, how='inner', lsuffix='_idh', rsuffix='_hiv')
```

## Join com os três indices, para cada ano separadamente:

Ano de 2000:

```python
aid_idz2 = aid_idx2.rename(index = str, columns = lambda column: 'aid_' + str(column))
```

```python
data2 = df2.join(aid_idz2, how="inner")
```

Ano de 2006:

```python
aid_idz = aid_idx.rename(index = str, columns = lambda column: 'aid_' + str(column))
```

```python
data = df.join(aid_idz, how="inner")
```

## Tirar o NaN:

Ano de 2000:

```python
Datta2 = data2.dropna(thresh = 3)
```

Ano de 2006:

```python
Datta1 = data.dropna(thresh = 3)
```

## Join com os data frames dos dois anos:

```python
Datta = Datta2.join(Datta1, how="inner")
```

## Salvar:

Salvando o data frame, execl e csv:

```python
Datta.to_excel('Projeto 1.xlsx', index = True)
```

```python
Datta.to_csv('Projeto 1.csv', index = True)
```

### Reta de tentendia linear para o ajuste dos dados:

Equação da Reta, que será utlizada após plotar os gráficos:

```python
def reta(dados_x, dados_y, ano, eixox, eixoy):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()
 
    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
 
    fig = plt.figure(figsize=(4, 3))
    plt.scatter(dados_x, dados_y, c='red', alpha=0.8)
    plt.title(ano)
    plt.xlabel(eixox)
    plt.ylabel(eixoy)
 
    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color='blue')
 
    plt.tight_layout()
    plt.show()
```

# Relatório:

Para iniciar a analise exploratória de dados de todos os paises, foram selecionados dois indices, e assim verificar se há alguma relação com o IDH ((Índice de Desenvolvimento Humano), que é medido através da expectativa de Vida, anos médios de estudo e anos esperados de escolaridade e a renda (GDP ou PIB) per capita. Os indices selecionados foram a prevalência do HIV nas pessoas de 15 a 49 anos e a ajuda recebida pelo comite de Ajuda ao Desenvolvimento da Organização para a Cooperação e Desenvolvimento Económico (CAD/OCDE). A análise foi direcionada em dois, 2000 e 2006, para assim responder a seguinte pergunta: “Considerando o aumento do Idh no de 2000 para 2006 na maioria dos paises, é correto afirmar que os indices de prevalência do vírus HIV e o índice de assistência oficial ao desenvolvimento de cada país influenciam o IDH?”
	Ao selecionar os indices no site https://www.gapminder.org , e os anos, a biblioteca pandas foi utilizada para ler os arquivos em Excel, logo depois, para unir os dados de todos os indices em um único Data Frame, foi necessário fazer a mudança dos indices das tabelas, isso foi feito separadamente para cada ano, obtendo dois datas frames, df e df2, para o ano de 2006 e 2000 respectivamente, o comando “join” foi usado novamente para unir os dois datas frames em um só chamado Datta, logo após esse data frame foi salvado como arquivo .xlsx e .csv.
	Para iniciar a plotagem dos gráficos e por fim analisa-los, a função da reta de ajuste com tendência linear para o gráfico de dispersão foi obtida na aula. Vários gráficos foram plotados a partir da matriz de dispersão que possibilita a visualização de todos os indices relacionados em gráficos. Com isso foi possível verificar que era necessário utilizar o log para garantir um melhor ajuste da reta.
	É possível verificar no primeiro gráfico de dispersão, que há uma associação negativa entre o auxílio de desenvolvimento (AID recive) e o IDH no ano de 2000, o mesmo ocorre no ano de 2006, com parando os mesmos indices a associação permanece negativa sem alterar significamente a inclinação, como foi possível ver no cor, para o ano de 2000 o corr foi de -0,68 e no ano de 2006 o cor foi de -0,73. Já comparando o índice de prevalência do HIV com o IDH, os dados se encontram muito dispersos, assim a reta de ajuste não demostra a melhor acomodação dos dados, apesar disso, a reta de “ajuste” teve uma associação negativa em ambos os anos, com o corr de -0,46 em 2000 e -0,50 em 2006.
	Com isso foi concluído que os paises subdesenvolvidos que recebem o auxílio do comitê de ajuda ao desenvolvimento, continuam sendo os paises com o menor índice de desenvolvimento humano, e também que possuem a maior prevalência do vírus. É possível afirmar também que devido a associação negativa na relação entre o IDH e AID nos anos de 2000 e 2006, receber o auxílio não é o bastante para que o país consiga se desenvolver de maneira eficaz.
	Em trabalhos futuros pode ser investigado indices que são diretamente relacionados com a prevalência do vírus HIV e com o índice de assistência, para assim determinar o porquê de tal ação não ser eficaz.
